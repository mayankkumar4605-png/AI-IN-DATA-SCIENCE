```python
#Find factorial using function
def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact = fact * i
    return fact

print("Factorial =", factorial(5))
```

```python
#check prime and not prime using function user input number
def check_prime(num):
    if num <= 1:
        return "Not Prime"

    for i in range(2, num):
        if num % i == 0:
            return "Not Prime"

    return "Prime"


# user input
n = int(input("Enter a number: "))

result = check_prime(n)
print("The number is", result)
```
