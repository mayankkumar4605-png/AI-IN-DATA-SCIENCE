```python
#Data points with the help of loop
x = [2,3,4,5,6]
y = [6,7,8,5,4]
for i in range(len(x)):
    print("Data Point", i+1, ":", x[i], y[i])
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y, marker='o')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Data Points using Loop")
plt.show()
```

```python
#Two Line Single Graph
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]
y2 = [5, 15, 25, 35, 45]

for i in range(len(x)):
    print("Line 1 Data Point:", x[i], y1[i])
    print("Line 2 Data Point:", x[i], y2[i])

plt.plot(x, y1, marker='o', label='Line 1')
plt.plot(x, y2, marker='s', label='Line 2')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Two Lines in a Single Graph")
plt.legend()
plt.show()
```

```python
#Add legend in graph
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]
y2 = [5, 15, 25, 35, 45]

plt.plot(x, y1, marker='o', label='Line 1')
plt.plot(x, y2, marker='s', label='Line 2')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Two Lines in One Graph")
plt.legend()      # This adds the legend
plt.show()
```

```python

plt.figure(figsize=(8, 6))

# First subplot
plt.subplot(2, 1, 1)   # (rows, columns, position)
plt.plot(x, y1, marker='o', label='Line Chart 1')
plt.title("First Line Chart")
plt.legend()

# Second subplot
plt.subplot(2, 1, 2)
plt.plot(x, y2, marker='s', label='Line Chart 2')
plt.title("Second Line Chart")
plt.legend()

plt.tight_layout()
plt.show()
```

#Bar Chart

```python
import matplotlib.pyplot as plt

name = ["Mayank","Parijat","Ankit","Aayush","Ujjwal"]
marks = [34,56,78,67,45]

plt.figure(figsize=(10,5))
plt.bar(name,marks,color='g',width=0.3)
plt.xlabel("Name")
plt.ylabel("Marks")
plt.legend("marks")
plt.bar(name,marks,color=["red","yellow","green","blue"])
plt.grid()
plt.scatter(name,marks)
for index,value in enumerate(marks):
    plt.text(index,value,str(value))
plt.show()
```

#Scatter Plot



```python
age = [45,65,49,67,47]
salary = [400,350,2000,3445,5000]
bubble_size = [500,700,800,900,1000]
color = ['red',"yellow","green","blue","black"]
plt.figure(figsize=(10,5))
plt.scatter(age,salary)
plt.grid()
plt.xlabel("age")
plt.ylabel("salary")
plt.title("age vs salary")
plt.scatter(age,salary,bubble_size,color,alpha=0.5,edgecolor="r")
plt.colorbar(label="intensityofcolor")
for index,value in enumerate(salary):
    plt.text(age[index],value,str(value))

plt.show()
```

#Line chart Practice

```python
sales ={
    "Months" : ["jan","feb","mar","apr","may","jun","jul","aug","sep","oct","nov","dec"],
    "Sales_unit": (22,24,26,28,30,32,34,36,38,40,42,44),
    "Revenue_in_lakhs" : (2,3,4,5,6.4,4.5,6.3,7,7.8,8.2,9,10)
}
import matplotlib.pyplot as plt

plt.plot(sales["Months"], sales["Sales_unit"],color = "r", marker = '*')

plt.grid()
plt.xlabel("Months")
plt.ylabel("Revenue<lakhs>")
plt.title("Months Vs Revenue")
plt.show()
```

```python
sales ={
    "Months" : ["jan","feb","mar","apr","may","jun","jul","aug","sep","oct","nov","dec"],
    "Sales_unit": (22,24,26,28,30,32,34,36,38,40,42,44),
    "Revenue_in_lakhs" : (2,3,4,5,6.4,4.5,6.3,7,7.8,8.2,9,10)
}
import matplotlib.pyplot as plt

plt.bar(sales["Months"], sales["Sales_unit"], color = "r")

plt.grid()
plt.legend("Sales_unit")
plt.xlabel("Months")
plt.ylabel("Sales_unit")
plt.title("Months Vs Sales Unit")
plt.show()
```

```python
plt.hist(sales["Sales_unit"])
plt.show()
```

```python
q1 = sum(sales["Revenue_in_lakhs"][0:3])
q2 = sum(sales["Revenue_in_lakhs"][3:6])
q3 = sum(sales["Revenue_in_lakhs"][6:9])
q4 = sum(sales["Revenue_in_lakhs"][9:12])
mayank = [q1,q2,q3,q4]

labels = ["Q1 ", "Q2 ", "Q3 ", "Q4"]

plt.pie(mayank, labels=labels, autopct="%.1f%%", startangle=4)

plt.title("Quarterly Revenue Distribution")
plt.legend()

plt.show()
```
