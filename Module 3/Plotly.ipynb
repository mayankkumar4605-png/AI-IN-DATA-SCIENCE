#Data visualization with plotly

```python
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_India = px.data.gapminder().query("country == 'India'")
data_India
graph =px.bar(data_India, x='year', y='pop', title='Population of India', labels={'pop':'Population'}, color='pop',hover_data=['pop'],text = 'pop')
graph.update_traces(texttemplate='%{text:.2s}', textposition='inside')
graph.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')
graph.update_layout(xaxis_title='Year', yaxis_title='Population',yaxis_tickangle = -45, template = "plotly_dark")
graph.show()
```

#Line chart use Plotly

```python
data_India = px.data.gapminder().query("country == 'India'")
data_India
graph =px.line(data_India, x='year', y='pop', title='Population of India', labels={'pop':'Population'}, color='pop',hover_data=['pop'],text = 'pop')
graph.update_traces(texttemplate='%{text:.2s}', textposition='top center')
graph.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')
graph.update_layout(xaxis_title='Year', yaxis_title='Popula0tion',yaxis_tickangle = -45, template = "plotly_dark")
graph.show()
```

#Scatter plot

```python
graph =px.scatter(data_India, x='gdpPercap', y='lifeExp', size = "pop", labels = {"pop":"Population"},color ="pop",hover_data = ["pop"],text = "pop")
graph.update_traces(texttemplate='%{text:.2s}', textposition='top center')
graph
```

#Pei Chart

```python
graph =px.pie(data_India, names='year', values='pop', title='Population Distribution of India by Year', labels={'pop':'Population'}, color='pop',hover_data=['pop'],hole = 0.5)
graph.update_traces(texttemplate='%{text:.2s}', textposition='inside')
graph
```

```python
graph =px.histogram(data_India, x='year',color = "pop")
graph
```

#Heat map

```python
graph =px.density_heatmap(data_India, x='year', y='pop', title='Population Distribution of India by Year', labels={'pop':'Population'}, color_continuous_scale="Viridis")
graph
```

#Box plot

```python
graph =px.box(data_India,y='lifeExp',points = "all",color_discrete_sequence= ["red"])
graph.update_traces(marker=dict(opacity=0.0))
graph
```
