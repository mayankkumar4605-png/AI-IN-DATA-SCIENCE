#Data visualization using Seaborn


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv('penguins.csv')
df.head()
```

#Line plot

```python
from pandas.io.formats import style
sns.lineplot(x='bill_length_mm', y='bill_depth_mm', data=df,hue = "sex",palette="BuPu",markers = ["o","s"],style = "sex",dashes= True,legend = "full")
sns.set_theme(style ="ticks")
plt.show()
```

#Scatter plot

```python
sns.scatterplot(data=df, x='bill_length_mm', y='bill_depth_mm', hue='species')
plt.figure(figsize=(10,10))
sns.set_theme(style='darkgrid',context="notebook")
plt.show()
```

#Line plot

```python
sns.lineplot(x='bill_length_mm', y='bill_depth_mm', data=df, hue='species')
plt.figure(figsize=(10,10))
sns.set_style('darkgrid')
plt.show()
```

#Histogram plot

```python
sns.histplot(data=df, x='bill_length_mm', hue='species')
plt.figure(figsize=(10,10))
sns.set_style('darkgrid')
plt.show()
```

#Box plot

```python
sns.boxplot(data=df, x='species', y='bill_length_mm')
plt.figure(figsize=(10,10))
plt.show()
```

#Violinplot

```python
sns.violinplot(data=df, x='species', y='bill_length_mm')
plt.figure(figsize=(10,10))
plt.show()
```

#Barplot

```python
sns.barplot(data=df, x='species', y='bill_length_mm')
plt.figure(figsize=(10,10))
plt.show()
```

#count plot

```python
sns.countplot(data=df, x='species')
plt.figure(figsize=(10,10))
plt.show()
```
