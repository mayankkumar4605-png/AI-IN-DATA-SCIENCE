#Merge in DataFrame

```python
#import libraries
import pandas as pd

#Creat DF's
df1=pd.DataFrame({"A":[80,85,7],"B":[1,4,3]})
df2=pd.DataFrame({"A":[80,85,8],"B":[1,4,3]})

#perform merge
pd.merge(df1,df2,on="A")
```

#USING HOW

```python
#INNER
df1=pd.DataFrame({"A":[80,85,7],"B":[1,4,3]})
df2=pd.DataFrame({"A":[80,85,8],"C":[5,6,8]})

#perform merge
pd.merge(df1,df2,on="A",how="inner")
```

```python
#INNER
df1=pd.DataFrame({"A":[80,85,7],"B":[1,4,3]})
df2=pd.DataFrame({"A":[80,85,8],"C":[5,6,8]})

#perform merge
pd.merge(df1,df2,on="A",how="outer")
```

```python
#INNER
df1=pd.DataFrame({"A":[80,85,8,9],"B":[1,2,3,4]})
df2=pd.DataFrame({"A":[80,85,8,10],"B":[1,2,3,5]})

#perform merge
pd.merge(df1,df2,on="A",how="inner")
```

```python
#OUTER
df1=pd.DataFrame({"A":[80,85,8,9],"B":[1,2,3,4]})
df2=pd.DataFrame({"A":[80,85,8,10],"B":[1,2,3,5]})

#perform merge
pd.merge(df1,df2,on="A",how="outer")
```

```python
#LEFT
df1=pd.DataFrame({"A":[80,85,8,9],"B":[1,2,3,4]})
df2=pd.DataFrame({"A":[80,85,8,10],"B":[1,2,3,5]})

#perform merge
pd.merge(df1,df2,on="A",how="left")
```

```python
#Right
df1=pd.DataFrame({"A":[80,85,8,9],"B":[1,2,3,4]})
df2=pd.DataFrame({"A":[80,85,8,10],"B":[1,2,3,5]})

#perform merge
pd.merge(df1,df2,on="A",how="right")
```

#INDEXING

```python
#Indexing
df1=pd.DataFrame({"A":[80,85,8,9],"B":[1,2,3,4]})
df2=pd.DataFrame({"A":[80,85,8,9],"B":[6,6,9,5]})

#perform merge
pd.merge(df1,df2,right_index=True,left_index=True,suffixes=("_id","_rollno"))
```

```python
#Merge 3 dataframe in Pandas
import pandas as pd

df1 = pd.DataFrame({"A":[23,24,25],"B":[12,13,14]})
df2 = pd.DataFrame({"A":[23,24,25],"B":[12,13,14]})
df3 = pd.DataFrame({"A":[23,24,25],"B":[12,13,14]})

# Perform sequential merges
merged_df_1_2 = pd.merge(df1, df2, on="A", suffixes=('_df1', '_df2'))
merged_df = pd.merge(merged_df_1_2, df3, on="A", suffixes=('', '_df3'))

print(merged_df)
```

```python
#Validate key(one to one)
import pandas as pd

#Creat dataframe
df1=pd.DataFrame({"A":[80,85,7],"B":[1,4,3]})
df2=pd.DataFrame({"A":[80,85,8],"B":[1,4,3]})

#perform merge
pd.merge(df1,df2,on="A",validate="one_to_one")
```

```python
#Validate key(one to many)
import pandas as pd

#Creat dataframe
df1=pd.DataFrame({"A":[80,85,7],"B":[1,4,3]})
df2=pd.DataFrame({"A":[80,85,8],"B":[1,4,3]})

#perform merge
pd.merge(df1,df2,on="A",validate="one_to_many")
```

```python
#Validate key(many to one)
import pandas as pd

#Creat dataframe
df1=pd.DataFrame({"A":[80,85,7],"B":[1,4,3]})
df2=pd.DataFrame({"A":[80,85,8],"B":[1,4,3]})

#perform merge
pd.merge(df1,df2,on="A",validate="many_to_one")
```

```python
#Validate key(many to many)
import pandas as pd

#Creat dataframe
df1=pd.DataFrame({"A":[80,85,7],"B":[1,4,3]})
df2=pd.DataFrame({"A":[80,85,8],"B":[1,4,3]})

#perform merge
pd.merge(df1,df2,on="A",validate="many_to_many")
```
