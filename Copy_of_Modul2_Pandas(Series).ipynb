**Data Manipulation with Pandas**

```
#import libraries
import pandas as pd
```

```
#creating a list
my_list=[1,2,3,4,5]

#creating the series with the help of pandas function
var=pd.Series(my_list)

#print the series
var
```

```
#indexing in series

#creating the series with the help of pandas function
var=pd.Series(my_list,index=["A","B","C","D","E"])

#print the series
print(var)
```

```
#change the data type,
var=pd.Series(my_list,index=["A","B","C","D","E"],dtype="float")

#print the series
var
```

```
#change the data type,
var=pd.Series(my_list,index=["A","B","C","D","E"],dtype="float",name="My Series")

#print the series
var
```

```
#creating series with the help of dictionary
dict={

    "A":(1,2,3),
    "B":(2,3,4),
  }
var=pd.Series(dict)
var
```

```
#creating series with the help of dictionary
dict={

    "Name":["Karan","Mayank","Ayush"],
    "Marks":[34,23,24],

  }
var=pd.Series(dict)
var
```

```
#print single variable i series
single_var=pd.Series(5,index=["A","B","C"])
single_var
```

#Operations in Pandas Series

```
#Opreation in series

#creating first list
num1=pd.Series(10,index=[1,2,3,4])

#creating second list
num2=pd.Series(5,index=[1])

#Addition operation in series
num1+num2
```

```
#Opreation in series

#creating first list
num1=pd.Series(5,index=[1,21,3,4])

#creating second list
num2=pd.Series(50,index=[51,6,7,8])

#Addition operation in series
num1+num2
```

```
import pandas as pd

#Opreation(+,-,*,?) in series

#creating first list
num1=pd.Series([1,2,3,4])

#creating second list
num2=pd.Series([5,6,7,8])

#Addition operation in series
num1+num2
```

#INSERT IN PANDAS


```
import pandas as pd

# Define the initial dictionary
info = {
    "Name": pd.Series(["mayank","Dhruv","ayush"]),
    "Marks": pd.Series([45,34,53])
}

info = pd.DataFrame(info)

age_values = [25, 30, 22]

info.insert(2, 'Age', age_values)

print(info)
```
