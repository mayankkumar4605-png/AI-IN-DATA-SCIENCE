#handle Missing Values

```python
df['Years of Experience'] = df['Years of Experience'].fillna(df['Years of Experience'].mean())
```

```python
df
```

```python
#Fill missing category values
df['Location'] = df['Location'] .fillna("Unknown")
```

```python
df
```

```python
#Remove duplicate rows
df=df.drop_duplicates()

#Check the fix datatype
df.dtypes

```

```python
df['Age Group'] = df['Age Group']
df
```

```python
#Rename column(optional)
df = df.rename(columns={
 'Years of Experience': 'YOF',
 'Location': 'Loc'
})
df
```

```python
# Handle Outliers (Simple Filtering)
df_sort = df[df['Monthly Salary (INR)'] < 10000]
df_sort.head()
{"repr_error":"0","type":"dataframe","variable_name":"df_sort"}
```

```python
#Clean text data
df['Education'] = df['Education'].str.strip()
#Converts all city names to lowercase letters.
df['Loc'] = df['Loc'].str.lower()

df
```

#Data Validation


```python
df = df[df['YOF'] > 0]
#df = df[(df['Age'] > 0) & (df['Age'] <= 120)]
df
```
